```python
import json
from typing import Optional

import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
import seaborn as sns
import yaml
from cmcrameri import cm
from matplotlib.colors import ListedColormap


with open("../../plotting_config.yaml", "r") as fp:
    config = yaml.safe_load(fp)

matplotlib.rcParams["svg.hashsalt"] = "42"


def calculate_gini(array: np.array) -> np.float64:
    """based on: https://github.com/oliviaguest/gini (CC0 1.0)"""
    # Values must be sorted:
    array = np.sort(array)
    # Index per array element:
    index = np.arange(1, array.shape[0] + 1)
    # Number of array elements:
    n = array.shape[0]
    # Gini coefficient:
    return np.round((np.sum((2 * index - n - 1) * array)) / (n * np.sum(array)), 4)


def calculate_multimodal_mean_pp(data: dict[str, float]) -> dict[str, Optional[float]]:
    result = {}
    for k, v in data.items():
        if len(v) > 0:
            result[k] = round(np.mean(v), 3)
        else:
            result[k] = None
    return result


def calculate_multimodal_mean_pp_as_dataframe(data: dict[str, float]) -> pd.DataFrame:
    result = calculate_multimodal_mean_pp(data)
    return pd.DataFrame(
        {
            "stop_id": list(result.keys()),
            "mean_price": list(result.values()),
        }
    )


colors = ["#41afaa", "#466eb4", "#00a0e1", "#e6a532", "#d7642c", "#af4b91"]
```

```python
points_in_bp = pp[pp["geometry"].intersects(bp.geometry[0])]
points_in_area = sgfw_raw.sjoin(pp).drop_duplicates("index_right")

print(len(pp))
print(points_in_bp.shape[0])
print(points_in_area.shape[0])
```

```python
labels = [
    f"${f:0.2f}-{t:0.12}$"
    for f, t in zip(np.arange(0.25, 2.6, 0.25)[:-1], np.arange(0.25, 2.6, 0.25)[1:])
]
filtered_pp = points_in_area.copy()
if FILTER:
    filtered_pp = points_in_area[points_in_area["norm_price"].between(0.25, 2.5)].copy()
filtered_pp["Real estate price"] = pd.cut(
    filtered_pp["norm_price"],
    bins=np.arange(0.25, 2.6, 0.25),
    labels=labels,
    right=False,
    include_lowest=True,
)
```

```python
q1 = filtered_pp.describe().loc["25%", "norm_price"]
q2 = filtered_pp.describe().loc["50%", "norm_price"]
q3 = filtered_pp.describe().loc["75%", "norm_price"]
```

```python
fig, ax = plt.subplots(figsize=(8, 5), layout="constrained")
sns.histplot(
    filtered_pp.sort_values("norm_price"),
    x="norm_price",
    # palette="YlGnBu",
    palette=list(cm.acton_r.resampled(9).colors),
    edgecolor=".3",
    linewidth=0.5,
    bins=np.arange(0.25, 1.25, 0.05),
    hue="Real estate price",
    alpha=1,
    ax=ax,
)
ax.margins(0, 0.05)
ax.set_xlabel("Price of 1 m$^2$ real estate in 2018 (million HUF)", fontsize=14)
ax.set_ylabel("Count", fontsize=14)
vline_color = "#d7642c"
vline_color = "#2d2d2d"
ax.axvline(q1, color=vline_color, linestyle="dashed")
ax.axvline(q2, color=vline_color, linestyle="solid")
_ = ax.axvline(q3, color=vline_color, linestyle="dashed")
for i in ["png", "svg", "pdf"]:
    fig.savefig(
        f"../../output/figures/property_price_distribution.{i}",
        pad_inches=0,
        metadata=config["metadata"][i],
        bbox_inches="tight",
    )
```

```python
with open(f"../../data/stops/{CITY}/{VERSION}/accessible_stops.json", "r") as fp:
    accessible_stops = json.load(fp)
```

```python
result = {}
for stop, accessible in accessible_stops.items():
    result[stop] = sgfw[sgfw["stop_id"].isin(accessible)]["norm_price"].tolist()
```

```python
with open(f"../../output/{CITY}/{VERSION}/multimodal_pp.json", "w") as fp:
    json.dump(result, fp)
```

```python
multimodal_pp = calculate_multimodal_mean_pp_as_dataframe(result).sort_values("stop_id")
multimodal_pp["mean_price"] = round(multimodal_pp["mean_price"], 3)
# multimodal_pp.to_csv(
#     f"../../output/{CITY}/{VERSION}/multimodal_property_price.csv",
#     index=False,
#     float_format="%0.02f",
# )
```

```python
stop_gini = {}
for k, v in result.items():
    stop_gini[k] = calculate_gini(np.array(v))
```

```python
with open(f"../../output/{CITY}/{VERSION}/multimodal_stop_gini.json", "w") as fp:
    json.dump(stop_gini, fp)
```

```python
records = []
for k, v in stop_gini.items():
    records.append([k, v])
multimodal_gini = pd.DataFrame.from_records(
    records, columns=["stop_id", "multimodal_gini"]
).sort_values("stop_id")
multimodal_gini.to_csv(
    f"../../output/{CITY}/{VERSION}/multimodal_stop_gini.csv", index=False
)
```

```python
walk_accessibility = gpd.read_file(f"../../output/{CITY}/{VERSION}/isochrones.geojson")
walk_accessibility = walk_accessibility.query("costing == 'walk' & range == 15").copy()
walk_accessibility = walk_accessibility.sjoin(pp).drop("index_right", axis=1)
```

```python
walk15_pp = walk_accessibility.groupby("stop_id")["norm_price"].mean().reset_index()
walk15_pp.rename(columns={"norm_price": "mean_price"}, inplace=True)
walk15_pp["mean_price"] = round(walk15_pp["mean_price"], 3)
# walk15_pp.to_csv(
#     f"../../output/{CITY}/{VERSION}/walk_property_price.csv",
#     index=False,
#     float_format="%0.02f",
# )
```

```python
walk15_gini = (
    walk_accessibility.groupby("stop_id")["norm_price"]
    .apply(calculate_gini)
    .reset_index()
)
walk15_gini.rename({"norm_price": "walk15_gini"}, axis=1, inplace=True)

walk15_gini.sort_values("stop_id").to_csv(
    f"../../output/{CITY}/{VERSION}/walk15_stop_gini.csv",
    index=False,
)
```

```python
stop_socioecon = (
    walk15_gini.merge(multimodal_gini, on="stop_id")
    .merge(walk15_pp.rename(columns={"mean_price": "walk15_mean_price"}), on="stop_id")
    .merge(
        multimodal_pp.rename(columns={"mean_price": "multimodal_mean_price"}),
        on="stop_id",
    )
)
stop_socioecon.to_csv(
    f"../../output/{CITY}/{VERSION}/stop_socioecon_property_price.csv",
    index=False,
)
```
